```python
#splitting the specifications
phones_data['RAM'],phones_data['ROM'],phones_data['Expandable'] = phones_data['SPECIFICATION'].str.split('|',3).str
del phones_data['SPECIFICATION']
phones_data.head()
```

```python
phones_data.isnull().sum()
```

```python
del phones_data['Expandable']
phones_data
```

```python
#data cleaning
phones_data = phones_data.replace(to_replace='GB RAM', value='',regex=True)
phones_data
```

```python
phones_data = phones_data.replace(to_replace='32 MB RAM', value='1',regex=True)
phones_data
```

```python
phones_data = phones_data.replace(to_replace='128 GB ROM', value='4',regex=True)
phones_data
```

```python
phones_data.info()
```

```python
#changing the datatype of RAM
phones_data['RAM'] = phones_data['RAM'].astype('int')
```

```python
phones_data.info()
```

```python
phones_data.describe()
```

```python
#  DATA visualization, the price versus the ram

plt.figure(figsize=(10,6))
sns.barplot(x=phones_data['PRICE'], y=phones_data['RAM'])
plt.title('PRICE VS RAM')
```
